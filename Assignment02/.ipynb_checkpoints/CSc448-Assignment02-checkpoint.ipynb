## Assignment 02

In this notebook, I applied the gradient descent algorithm on the data from 'data.csv' to find its linear regression.

### Setups
In order to run this notebook, you need to install following modules
- numpy
- pandas
- matplotlib

```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

### Import Data
The cell below imports data from 'data.csv' and plot the data

```python
fig, ax = plt.subplots()

data = pd.read_csv('data.csv') # Preprocessing Input data
X = data.iloc[:, 0] # 1st col of data 
Y = data.iloc[:, 1] # 2nd col of data
ax.scatter(X, Y)
plt.show()
```

### Implementaion of Gradient Descent Algorithm
By using Gradient Descent Algorithm, the code will predict the slope m and y-intercept for data imported from 'data.csv'.
Initially, m and c are set to 0 and learn rate is set to 0.0001.
For each iteration, the gradient descent algorithm are applied to data, the m and c are updated. The predict value of m and c are stored in the list to use to generate gif that demonstrates the change of linear regression step by step.

```python
m = 0
c = 0
M = [0]
C = [0]
L = 0.0001  # The learning Rate

# Performing Gradient Descent 
for i in range(20): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    M.append(m)
    C.append(c)
    
print (m, c)
```

### Generate gif

The code use the stored value of m and c to demonstrates the change of the predict value of m and c.
The FuncAnnimation use the update function to generate the linear regression from beginning to end.

```python
line, = ax.plot(X, M[0]*X+C[0])
def update(i):
    label = 'm = {0}  c= {1}'.format(M[i], C[i])
    # Update the line and the label
    line.set_ydata(M[i]*X+C[i])
    ax.set_xlabel(label)
    return line, ax

anim = FuncAnimation(fig, update, frames=range(0, 20), interval=500)
anim.save('line.gif', dpi=80, writer='pillow')
```
