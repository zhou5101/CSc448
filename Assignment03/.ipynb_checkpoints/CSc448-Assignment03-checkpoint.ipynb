## Abstract

Gold is a special currency used for reserves and investment, it is also an important material for the jewelry industry, electronics industry, modern communications, aerospace industry and other sectors. The gold price is closely related with the value of U.S. dollars, by predicting the gold price, we can know the current value of the U.S. dollar. I will use the past data of gold price to train a Linear Regression to predict the future gold price.

Mushroom is a spore-bearing fruiting body of a fungus, typically produced above ground, on soil, or on its food source. Tasting and smelling mushrooms carries its own hazards because of poisons and allergens. Therefore it is essential to identify the edible and poisonous mushrooms. I will use a k-Nearest Neighbors to classify these two types mushrooms based on their features.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

## Section 1: Regression Dataset Prep


In the gold price dataset, it contains the past gold and silver price in U.S. dollar and EUR, also the stock price related to gold. I will mainly use the data gold price in U.S. dollar for Linear Regression.

```python
gold = pd.read_csv('./gld_price_data.csv')
gold.head(15)
```

Since the only data in this Linear Regression is gold price, and next day gold price is primarly based on the past gold price. Therefore, I used the mean of past 3 days gold price as independent varible, the gold price for next day as dependent variable.
1. S_3 - the mean of past three days gold price
2. Next_Day_Price - the gold price for next day

```python
gold['S_3'] = gold['GLD'].rolling(window=3).mean()
gold['Next_Day_Price'] = gold['GLD'].shift(-1)
gold = gold.dropna()
gold = gold[['Date','GLD', 'S_3', 'Next_Day_Price']]
gold.head(15)
```

## Section 2: Linear Regression

```python
X = gold[['S_3']]
Y = gold['Next_Day_Price']
linear = LinearRegression()
```

I will reverse 30% of the dataset for testing, and choose not to shuffle the dataset.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, shuffle=False)
Date_train, Date_test = train_test_split(gold['Date'], test_size=0.3, shuffle=False)
linear = linear.fit(X_train, Y_train)
```

```python
Y_pred = linear.predict(X)
Y_pred = pd.DataFrame(Y_pred, columns=['Next_Day_Price'])
gold['Date'].index = np.arange(0, len(Y_pred))
dates = pd.DataFrame(gold['Date'], columns=['Date'])
Y_pred = pd.concat([dates,Y_pred], axis=1)
Y_pred['Date'] = pd.to_datetime(Y_pred['Date'])

Y_true_train = pd.DataFrame(Y_train, columns=['Next_Day_Price'])
date_true_train = pd.DataFrame(Date_train, columns=['Date'])
Y_true_train = pd.concat([date_true_train, Y_true_train], axis=1)
Y_true_train['Date'] = pd.to_datetime(Y_true_train['Date'])

Y_true_test = pd.DataFrame(Y_test, columns=['Next_Day_Price'])
date_true_test = pd.DataFrame(Date_test, columns=['Date'])
Y_true_test = pd.concat([date_true_test, Y_true_test], axis=1)
Y_true_test['Date'] = pd.to_datetime(Y_true_test['Date'])

ax = Y_true_train.plot(x='Date', y='Next_Day_Price')
Y_true_test.plot(ax=ax, x='Date', y='Next_Day_Price', figsize=(12, 9))
Y_pred.plot(ax=ax, x='Date', y='Next_Day_Price')
plt.legend(['Training Data', 'Testing Data', 'Predicted Price'])
plt.ylabel("Gold Price")
plt.title("Predicted Price VS Actual Price")
plt.show()
```

The training data is the blue line, the testing data is the organge line, the linear regression for the dataset is the green line. As shown in above figure, the prediction from linear regression almost covered the all training and testing data

```python
Y_predict_test = linear.predict(X_test)
Y_predict_test = pd.DataFrame(
    Y_predict_test, index=Date_test, columns=['Predicted Price'])

Y_test.index = Date_test
Y_test = pd.DataFrame(Y_test)
```

```python
r2_score = linear.score(Y_test, Y_predict_test)*100
scores = cross_val_score(linear, X, Y, cv=5)
r2_score 
```

```python
scores = cross_val_score(linear, X, Y, cv=5)
print(scores)
```

Based on the r square value and cross validation score, it seems like the linear regression model is a very effective predictor for the gold price. The correlation between the average of past three days gold price and the next day gold price since the score is above 95%. I believe if I add the more parameter for the past average gold price, the score will increase.

```python
df = pd.read_csv('./mushrooms.csv')
df.head(15)
```

Since I find a lot of values of stalk root are ? in this category, so I droped this column

```python
mushroom=df.drop(['stalk-root'],axis=1)

mappings = list()

encoder = LabelEncoder()

for column in range(len(mushroom.columns)):
    mushroom[mushroom.columns[column]] = encoder.fit_transform(mushroom[mushroom.columns[column]])
    mappings_dict = {index: label for index, label in enumerate(encoder.classes_)}
    mappings.append(mappings_dict)
    
```

```python
mushroom.head(15)
```

Here's a summary of the data mapping for each of the attribute used for the k-Nearest Neighbors:

```python
mappings
```

## Section 4: Classification

```python
X = mushroom.drop('class', axis=1)
Y = mushroom['class']
KNN = KNeighborsClassifier()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle =True, test_size=0.3, random_state=42)

KNN.fit(X_train, Y_train)

Y_pred = KNN.predict(X_test)

accuracy_score(Y_test, Y_pred)
```

```python
score = cross_val_score(KNN, X, Y, cv=10)
score
```

Based on the accuracy score, the k-Nearest Neighbors is a perfect effective predictor of edible and posionous mushrooms. Although the corss validation score does not show it is perfect, but for most dataset it rates 99%.
